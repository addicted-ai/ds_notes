## One vs One

- It involves 3 steps:
    - Create subsets
    - Model training
    - Model prediction
- In create subsets stage, a data set is divided into multiple subsets each containing two target variables. Hence, if there are 'n' target variables, we will have nC2 subsets.

- In the second step i.e model training, for each subset, a separate model is built and trained. 
- Since the data set has n target variables, we will have nC2 subsets in total, and, thus, we will build nC2 models.

- In the third step, which is model prediction, test samples are passed to **each of the nC2 models**, which predict/classify the test samples accordingly.
- After all the nC2 models have predicted class for test sample, the test sample gets classified under that target variable that has the **highest count**.
- If maximum count does not belong to one single class, but two or more target variables have the same count, then we will follow a different approach.
- Since there will be n models associated with any target, we will have n probability & prediction for them too. We'll just pick maximum probability associated with the target & maximum probability of 2nd target, then assign final class depending whichever class has **maximum probability**.

## One vs Rest

- It involves 3 steps:
    - Create Dataset
    - Model training
    - Model prediction
- In dataset creation part, we are not creating subsets, we are considering **all datapoints**.
- Total dataset created will be equal to total no of classes/target variables.
- Each dataset will have one single class labeled as 1, & rest classes labeled as 0.

- In model training part, we build models for all n dataset.
- In third step, we predict for all datapoints using n models. So for each datapoint, we'll have n predictions using n models & probabilities for n classes for single datapoint.
- Since for single datapoint, we'll have probability for all classes, we can pick maximum probability to assign final class.

## Logistic Equation

Softmax Function:
- class 0 probability: $$p_0 = \frac{e^{w^T_0x'}}{e^{w^T_0x'} + e^{w^T_1x'} + .... + e^{w^T_nx'}}$$
- class 1 probability: $$p_1 = \frac{e^{w^T_1x'}}{e^{w^T_0x'} + e^{w^T_1x'} + .... + e^{w^T_nx'}}$$
- class n probability: $$p_1 = \frac{e^{w^T_nx'}}{e^{w^T_0x'} + e^{w^T_1x'} + .... + e^{w^T_nx'}}$$

### Logistic function (limiting class to 2)

- class 0 probability: $$p_0 = \frac{e^{w^T_0x'}}{e^{w^T_0x'} + e^{w^T_1x'}}$$
- class 1 probability: $$p_1 = \frac{e^{w^T_1x'}}{e^{w^T_0x'} + e^{w^T_1x'}}$$

Simplifying class 1 probability by dividing $e^{w^T_1x'}$

$$p_1 = \frac{e^{w^T_1x'}}{e^{w^T_0x'} + e^{w^T_1x'}}
= \frac{1}{\frac{e^{w^T_0x'}}{e^{w^T_1x'}} + 1}
= \frac{1}{1 + e^{(w^T_0 - w^T_1)x'}}
$$
if we replace $w_1 - w_0 = w$, then
$$p_1 = \frac{1}{1 + e^{-w^Tx'}}$$


where
$$ w^Tx_i = 
\left[ w_1 w_2 w_3 . . w_k \right] .
\left[
\begin{matrix}
  x_{i1} \\

  . \\
  x_{ik}
\end{matrix}
\right]
= w_1x_{i1} + w_2x_{i2} + w_3x_{i3} + . . + w_kx_{ik}
 $$

### Binary cross-entropy loss:

For i<sup>th</sup> data point: $l_i(y_i;p) = -(y_i log(p_i) + (1 - y_i) log(1 - p_i))$

Total Loss: $L(y;p) = -1/n \sum_{i=0}^n y_i log(p_i) + (1 - y_i) log(1 - p_i)$

### Multi-class cross-entropy loss:

Total Loss: $$L(y;p) = -1/n \sum_{i=0}^n y log(p)$$
